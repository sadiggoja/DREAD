```python
consumption=pd.read_csv("powerconsumption.csv")
```

```python
consumption['Datetime'] = pd.to_datetime(consumption['Datetime'])
consumption.set_index('Datetime', inplace=True)
```

```python
consumption
```

```python
features = ['Temperature', 'Humidity', 'WindSpeed', 'GeneralDiffuseFlows', 'DiffuseFlows',
            'PowerConsumption_Zone1', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3']
df=consumption.copy()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[features])
```

```python
one_class_svm = OneClassSVM(nu=0.05)
# Fit the model on the normal data (inliers)
one_class_svm.fit(scaled_data)
```

```python
predictions = one_class_svm.predict(scaled_data)

# Convert predictions to 1 for normal, -1 for anomaly
predictions = (predictions == 1).astype(int)
```

```python
print((predictions == 0).sum())
print((predictions == 1).sum())
```

```python
df['anomaly_score']=predictions
```

```python
df
```

```python
plt.plot(df.index, df['PowerConsumption_Zone1'])
plt.scatter(df.index[df['anomaly_score'] == 0], df['PowerConsumption_Zone1'][df['anomaly_score'] == 0], c='red')
plt.xlabel('Date')
plt.ylabel('PowerConsumption_Zone1')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'gamma': [0.0001, 0.001, 0.01, 0.1, 1],
              'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],}
one_class=OneClassSVM(nu=0.05)
grid_search = GridSearchCV(estimator=one_class, scoring='neg_mean_squared_error', param_grid=parameters, cv=5)
grid_search.fit(scaled_data)
best_parameters = grid_search.best_params_
print(best_parameters)
```
