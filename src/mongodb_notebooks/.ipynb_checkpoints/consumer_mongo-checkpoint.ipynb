```python
from confluent_kafka import Consumer, KafkaException
import pandas as pd
import json


def msg_dict(data_string):
    # Split the string into key-value pairs
    key_value_pairs = data_string.split(',')

    # Create a dictionary from the key-value pairs
    data_dict = {}
    for pair in key_value_pairs:
        try:
            key, value = pair.split(':')
            data_dict[key] = value
        except:
            key, value = pair.split('e:')
            data_dict[key+'e'] = value

    return data_dict



def kafka_consumer_dataframe():
    consumer_conf = {'bootstrap.servers': 'localhost:9093', 
                     'group.id': 'processed-data-group00',
                     'auto.offset.reset': 'earliest' }
    c = Consumer(consumer_conf)
    c.subscribe(['topic_mongodb'])
    list_msg=list()
    try:
        while True:
            msg = c.poll(1.0)
            if msg is None:
                continue
            #if msg.error():
             #   if msg.error().code() == KafkaException._PARTITION_EOF:
              #      continue
               # else:
                #    print(msg.error())
                 #   break

            list_msg.append(msg_dict(msg.value().decode('utf-8')))
            
            if len(list_msg) % 1000 == 0:
                print(f"Rows Consumed: {len(list_msg)}")
    except KeyboardInterrupt:
        pass

    finally:
        c.close()
    print(f"Total:{len(list_msg)}")
    return list_msg
```

```python
stream=kafka_consumer_dataframe()
```

```python
from confluent_kafka import Consumer, KafkaException
import pandas as pd
from pymongo import MongoClient
```

```python

# Kafka consumer configuration
consumer_config = {'bootstrap.servers': 'localhost:9093', 
                     'group.id': 'processed-data-group00',
                     'auto.offset.reset': 'earliest' }

# MongoDB configuration
mongo_uri = 'mongodb://root:example@localhost:27016'  # Update with your MongoDB URI
mongo_db = 'admin'
mongo_collection = 'coll1'

# Create Kafka consumer
consumer = Consumer(consumer_config)

# Create MongoDB client and connect to the database
mongo_client = MongoClient(mongo_uri)
mongo_db = mongo_client[mongo_db]
mongo_collection = mongo_db[mongo_collection]

# Kafka topic
kafka_topic = 'topic_mongodb'
consumer.subscribe([kafka_topic])
```
